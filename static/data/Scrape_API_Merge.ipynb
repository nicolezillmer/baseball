```python
import pandas as pd
```

```python
pitching = pd.read_csv('munged_data/munged_pitcher.csv',encoding='utf-8')
batting = pd.read_csv('munged_data/munged_batting.csv',encoding='utf-8')
api =  pd.read_csv('api_data/munged_api.csv',encoding='utf-8')
```

```python
pitching
```

```python
batting
```

```python
api
```

```python
pitching_merge = pitching.merge(api, how= "left", on="PLAYER")
pitching_merge
```

```python
pitch = pitching_merge[pitching_merge.MLBAMID.isnull()]
pitch
```

```python
batting_merge = batting.merge(api, how= "left", on="PLAYER")
batting_merge
```

```python
bat = batting_merge[batting_merge.MLBAMID.isnull()]
bat
```

```python
pitch.to_csv('final_data/pitch_no_birthdate.csv', index=False)
bat.to_csv('final_data/bat_no_birthdate.csv', index=False)
```

```python
final_pitch = pitching_merge[pitching_merge.MLBAMID.notnull()]
final_bat = batting_merge[batting_merge.MLBAMID.notnull()]
final_pitch.to_csv('final_data/pitch_birthdate.csv', index=False)
final_bat.to_csv('final_data/bat_birthdate.csv', index=False)
```
