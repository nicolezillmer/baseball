```python
import pandas as pd
```

```python
# JSON from https://sportsdata.io/
df = pd.read_json('https://api.sportsdata.io/v3/mlb/scores/json/Players?key=6e47361b5deb4bf58f081aa3176c932e')
```

```python
df.to_csv('api_data/original_json.csv')
df
```

```python
df.columns
```

```python
df.drop(df.columns.difference([ 'MLBAMID', 'FirstName', 'LastName','BirthDate','Team', 'Jersey', 'BirthCity','BirthState', 'BirthCountry', 
                               'ProDebut']),1, inplace=True)
```

```python
df
```

```python
# take a slice of df for those with an official MLBAMID

mlb_players = df[df.MLBAMID.notnull()]
```

```python
mlb_players
```

```python
mlb = pd.DataFrame(mlb_players)
mlb['PLAYER'] = mlb["FirstName"] + " " + mlb["LastName"]
mlb
```

```python
mlb.columns
cols = ['MLBAMID', 'PLAYER','FirstName', 'LastName','BirthDate','Team', 'Jersey','BirthCity','BirthState', 'BirthCountry', 
                               'ProDebut']
```

```python
mlb=mlb[cols]
mlb
```

```python
# Filtering out players with a ProDebut
pro = mlb[mlb.ProDebut.notnull()]
pro
```

```python
pro.to_csv('api_data/munged_api.csv')
```
