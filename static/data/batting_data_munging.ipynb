```python
import pandas as pd
```

```python
batting = pd.read_csv('merged_data/UPDATED_merged_batting_data_10.13.2020.csv')
batting.columns
```

```python
# Drop all columns except the following

batting.drop(batting.columns.difference(['PLAYER','YRS00', 'YRS01', 'YRS02', 'YRS03', 'YRS04', 'YRS05',
       'YRS06', 'YRS07','YRS08', 'YRS09', 'YRS10', 'YRS11', 'YRS12', 'YRS13',
       'YRS14', 'YRS15','YRS16', 'YRS17', 'YRS18', 'YRS19']),1, inplace=True)
```

```python
batting.head()
```

```python
batting['count'] = batting.count(axis=1)
```

```python
# players need to have played at least 5 years within window
df = batting[batting['count']>= 6]
df
```

```python
batter_list = df.PLAYER.unique()
batter_list
```

```python
api_list = []
```

```python
for batter in batter_list:
    j = batter.replace(' ', '%25')
    api_list.append(j)
api_list
```

```python
api_df = pd.DataFrame(api_list)
api_df
```

```python
api_df.to_csv('player_list/batter_list.csv', index=False)
```

```python
batter_df =pd.DataFrame(batter_list)
batter_df.columns = ['PLAYER']
batter_df
```

```python
# call in original df without any altering and merge with filtered list
og_df= pd.read_csv('merged_data/UPDATED_merged_batting_data_10.13.2020.csv')
filtered_batter = pd.merge(batter_df,og_df,how='left', on='PLAYER')
filtered_batter
```

```python
no_year = filtered_batter.drop(['YRS00', 'YRS01', 'YRS02', 'YRS03', 'YRS04', 'YRS05',
       'YRS06', 'YRS07','YRS08', 'YRS09', 'YRS10', 'YRS11', 'YRS12', 'YRS13',
       'YRS14', 'YRS15','YRS16', 'YRS17', 'YRS18', 'YRS19'], axis=1)
no_year
```

```python
no_year.drop(no_year.columns.difference(['PLAYER','BA00', 'BA01', 'BA02', 'BA03', 'BA04', 'BA05',
       'BA06', 'BA07','BA08', 'BA09', 'BA10', 'BA11', 'BA12', 'BA13',
       'BA14', 'BA15','BA16', 'BA17', 'BA18', 'BA19']),1, inplace=True)
no_year
```

```python
no_year.to_csv('munged_data/munged_batting.csv', index=False)
```
