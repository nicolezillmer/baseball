```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime
from datetime import timedelta, date
```

```python
pitching = pd.read_csv('final_data/pitch_birthdate_final.csv',encoding='utf-8')
```

```python
pitching
```

```python
pitching.dtypes
```

```python
pitching['BirthDate'] = pd.to_datetime(pitching['BirthDate'])
pitching['ProDebut'] = pd.to_datetime(pitching['ProDebut'])
pitching
```

```python
def from_dob_to_age(born):
    today = datetime.date.today()
    return today.year - born.year - ((today.month, today.day) < (born.month, born.day))
```

```python
pitching['Age'] = pitching["BirthDate"].apply(lambda x: from_dob_to_age(x))
```

```python
pitching["AGE00"] = pitching["Age"]-20
pitching["AGE01"] = pitching["Age"]-19
pitching["AGE02"] = pitching["Age"]-18
pitching["AGE03"] = pitching["Age"]-17
pitching["AGE04"] = pitching["Age"]-16
pitching["AGE05"] = pitching["Age"]-15
pitching["AGE06"] = pitching["Age"]-14
pitching["AGE07"] = pitching["Age"]-13
pitching["AGE08"] = pitching["Age"]-12
pitching["AGE09"] = pitching["Age"]-11
pitching["AGE10"] = pitching["Age"]-10
pitching["AGE11"] = pitching["Age"]-9
pitching["AGE12"] = pitching["Age"]-8
pitching["AGE13"] = pitching["Age"]-7
pitching["AGE14"] = pitching["Age"]-6
pitching["AGE15"] = pitching["Age"]-5
pitching["AGE16"] = pitching["Age"]-4
pitching["AGE17"] = pitching["Age"]-3
pitching["AGE18"] = pitching["Age"]-2
pitching["AGE19"] = pitching["Age"]-1
```
