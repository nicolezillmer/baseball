```python
import pandas as pd
```

```python
pitching = pd.read_csv('merged_data/merged_pitch_data_10.12.2020.csv')
pitching.columns
```

```python
# Drop all columns except the following

pitching.drop(pitching.columns.difference(['PLAYER','YRS00', 'YRS01', 'YRS02', 'YRS03', 'YRS04', 'YRS05',
       'YRS06', 'YRS07','YRS08', 'YRS09', 'YRS10', 'YRS11', 'YRS12', 'YRS13',
       'YRS14', 'YRS15','YRS16', 'YRS17', 'YRS18', 'YRS19']),1, inplace=True)
```

```python
pitching.head()
```

```python
pitching['count'] = pitching.count(axis=1)
```

```python
# players need to have played at least 5 years within window
df = pitching[pitching['count']>= 6]
df
```

```python
pitcher_list = df['PLAYER'].tolist()
```

```python
api_list = []
```

```python
for pitcher in pitcher_list:
    j = pitcher.replace(' ', '%25')
    api_list.append(j)
api_list
```

```python
api_df = pd.DataFrame(api_list)
api_df
```

```python
api_df.to_csv('player_list/pitcher_list.csv', index=False)
```

```python
pitcher_df =pd.DataFrame(pitcher_list)
pitcher_df.columns = ['PLAYER']
pitcher_df
```

```python
# call in original df without any altering and merge with filtered list
og_df= pd.read_csv('merged_data/merged_pitch_data_10.12.2020.csv')
filtered_pitcher = pd.merge(pitcher_df,og_df,how='left', on='PLAYER')
filtered_pitcher
```

```python
no_year = filtered_pitcher.drop(['_x','YRS00', 'YRS01', 'YRS02', 'YRS03', 'YRS04', 'YRS05',
       'YRS06', 'YRS07','YRS08', 'YRS09', 'YRS10', 'YRS11', 'YRS12', 'YRS13',
       'YRS14', 'YRS15','YRS16', 'YRS17', 'YRS18', 'YRS19'], axis=1)
no_year
```

```python
no_year.drop(no_year.columns.difference(['PLAYER','whip00', 'whip01', 'whip02', 'whip03', 'whip04', 'whip05',
       'whip06', 'whip07','whip08', 'whip09', 'whip10', 'whip11', 'whip12', 'whip13',
       'whip14', 'whip15','whip16', 'whip17', 'whip18', 'whip19', 'ERA00', 'ERA01', 'ERA02', 'ERA03', 'ERA04', 'ERA05',
       'ERA06', 'ERA07','ERA08', 'ERA09', 'ERA10', 'ERA11', 'ERA12', 'ERA13',
       'ERA14', 'ERA15','ERA16', 'ERA17', 'ERA18', 'ERA19']),1, inplace=True)
no_year
```

```python
no_year.to_csv('munged_data/munged_pitcher.csv', index=False)
```
